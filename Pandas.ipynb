## = Copy, Shallow Copy, Deep Copy

````python
df_subset= df[['Product line', 'Date', 'Quantity', 'Payment']].head()
df_subset.sort_index(axis=1)
````

````python
equals= df_subset
````

````python
equals
````

````python
equals['Quantity']= [1, 2, 3, 4, 5]
equals
````

````python
df_subset
````

````python
equals['Test']= [11, 22, 33, 44, 55]
equals
````

````python
df_subset
````

````python
df_subset= df[['Product line', 'Date', 'Quantity', 'Payment']].head()
df_subset.sort_index(axis=1)
````

````python
shallow_copy= df_subset.copy(deep=False)
shallow_copy
````

````python
shallow_copy['Quantity']= [1, 2, 3, 4, 5]
shallow_copy
````

````python
df_subset
````

````python
shallow_copy['Test']= [11, 22, 33, 44, 55]
shallow_copy
````

````python
df_subset
````

````python
df_subset= df[['Product line', 'Date', 'Quantity', 'Payment']].head()
df_subset.sort_index(axis=1)
````

````python
deep_copy= df_subset.copy()
deep_copy
````

````python
deep_copy['Quantity']= [1, 2, 3, 4, 5]
deep_copy
````

````python
df_subset
````

````python
deep_copy['Test']= [11, 22, 33, 44, 55]
deep_copy
````

````python
df_subset
````

|  **= Copy** | **Shallow Copy** | **Deep Copy** |
| ----------- | ---------------- | ------------- |
|Changing value in an existing column in new dataframe will change the value in same column in old dataframe as well. | Changing value in an existing column in new dataframe will change the value in same column in old dataframe as well. | Changing value in an existing column in new dataframe **will not change** the value in same column in old dataframe as well.|
|Adding or removing a column in new dataframe **will reflect** in old dataframe as well| Adding or removing a column in new dataframe **will not reflect** in old dataframe | Adding or removing a column in new dataframe **will not reflect** in old dataframe|
| **Conclusion:** Whatever changes are made in new dataframe, will reflect in old dataframe as well. | **Conclusion:** Changes in value will impact the old dataframe but the structure will not | **Conclusion:** Any kind of change in new dataframe will not impact the old dataframe. |

## Selection or Slicing & Dicing in a Pandas DataFrame

### `loc` vs `iloc` vs `ix`

### Row Selection
```py
df[start : end]
```

### Selecting Using loc
```py
# List method
df.loc[list of rows , list of columns]

# Range method
df.loc[R_start:R_end , C_start:C_end]

# Boolean Methon
df.loc[Row with boolean condition, Column or list of columns]
```
